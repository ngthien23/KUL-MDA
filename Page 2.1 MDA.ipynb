```python
# static_plot.py

# import necessary packages
import matplotlib.pyplot as plt
import pandas as pd

# data
data_lst = pd.read_csv('C:/Users/User/Downloads/Max_noise_minute.csv')
# Print the column names
column_names = data_lst.columns
print(column_names)
data_lst.head()
data_lst['timestamp'] = pd.to_datetime(data_lst['timestamp'])

# create the figure and axis objects
fig, ax = plt.subplots()

# plot the data and customize
ax.plot(data_lst['timestamp'], data_lst['MP 01: Naamsestraat 35  Maxim'])
ax.set_xlabel('Day Number')
ax.set_ylabel('Noise level')
ax.set_title('Noise level on MP 01: Naamsestraat 35')
```

```python
# animated_line_plot.py
from random import randint

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# create empty lists for the x and y data
x = data_lst['timestamp']
y = data_lst['MP 01: Naamsestraat 35  Maxim']
# create the figure and axes objects
fig, ax = plt.subplots()

# function that draws each frame of the animation
def animate(i):
    pt = randint(1,9) # grab a random integer to be the next y-value in the animation
    x.append(i)
    y.append(pt)

    ax.clear()
    ax.plot(x, y)
    ax.set_xlim([0,])
    ax.set_ylim([0,110])
    
    # run the animation
ani = FuncAnimation(fig, animate, frames=20, interval=500, repeat=False)

plt.show()
```

```python

import plotly.graph_objects as go
import pandas as pd

import datetime

# Filter data for the desired time range (2022-2023)
# Convert 'timestamp' column to datetime format
data_lst['timestamp'] = pd.to_datetime(data_lst['timestamp'])
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2023, 12, 31)
filtered_data = data_lst[(data_lst['timestamp'] >= start_date) & (data_lst['timestamp'] <= end_date)]


# create figure
fig = go.Figure()



# Update plot sizing
fig.update_layout(
    width=800,
    height=900,
    autosize=False,
    margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
)

# Add dropdown
fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([

                dict(
                    args=[{"visible": [True, False, False, False, False, False, False, False]}, {"title": "MP 01: Naamsestraat 35  Maxim"}],# "type", "animation2.gif"],
                    label="MP 01: Naamsestraat 35  Maxim",
                    method="update",
                ),
                dict(
                    args=[{"visible": [False, True, False, False, False, False, False, False]},{"title": "MP 02: Naamsestraat 57 Xior"}],#"type", "animation.gif"],
                    label="MP 02: Naamsestraat 57 Xior",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, True, False, False, False, False, False]},{"title": "MP 03: Naamsestraat 62 Taste"}],#"type", "animation2.gif"],
                    label="MP 03: Naamsestraat 62 Taste",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, False, True, False, False, False, False]},{"title": "MP 04: His & Hears"}],#"type", "animation2.gif"],
                    label="MP 04: His & Hears",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, False, False, True, False, False, False]},{"title": "MP 05: Calvariekapel KU Leuven"}],#"type", "animation2.gif"],
                    label="MP 05: Calvariekapel KU Leuven",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, False, False, False, True, False, False]},{"title": "MP 06: Parkstraat 2 La Filosovia"}],#"type", "animation2.gif"],
                    label="MP 06: Parkstraat 2 La Filosovia",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, False, False, False, False, True, False]},{"title": "MP 07: Naamsestraat 81"}],#"type", "animation2.gif"],
                    label="MP 07: Naamsestraat 81",
                    method="update"
                ),
                dict(
                    args=[{"visible": [False, False, False, False, False, False, False, True]},{"title": "MP 08: bis - Vrijthof"}],#"type", "animation2.gif"],
                    label="MP 08: bis - Vrijthof",
                    method="update"
                ),
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

# Set title
fig.update_layout(title_text="Maximum noise per location")

# Add the animation traces to the figure
fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 01: Naamsestraat 35  Maxim'],
        mode='lines',
        name='MP 01: Naamsestraat 35  Maxim'
    )
)

fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 02: Naamsestraat 57 Xior'],
        mode='lines',
        name='MP 02: Naamsestraat 35  Maxim'
    )
)

fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 03: Naamsestraat 62 Taste'],
        mode='lines',
        name='MP 03: Naamsestraat 62 Taste'
    )
)

fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 04: His & Hears'],
        mode='lines',
        name='MP 04: His & Hears'
    )
)
fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 05: Calvariekapel KU Leuven'],
        mode='lines',
        name='MP 05: Calvariekapel KU Leuven'
    )
)
fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 06: Parkstraat 2 La Filosovia'],
        mode='lines',
        name='MP 06: Parkstraat 2 La Filosovia'
    )
)
fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 07: Naamsestraat 81'],
        mode='lines',
        name='MP 07: Naamsestraat 81'
    )
)
fig.add_trace(
    go.Scatter(
        x=filtered_data['timestamp'],
        y=filtered_data['MP 08: bis - Vrijthof'],
        mode='lines',
        name='MP 08: bis - Vrijthof'
    )
)

# Add annotation
fig.update_layout(
    annotations=[
        dict(text="Trace type:", showarrow=False,
        x=0, y=1.085, yref="paper", align="left")
    ]
)

fig.show()
```
